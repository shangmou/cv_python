# 直方图

![title](hist_1.png)

```python
def cv_show(img,name):
    cv2.namedWindow(name,0)
    cv2.imshow(name,img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

# 函数：cv2.calcHist(images,channels,mask,histSize,ranges)

- images: 原图像图像格式为 uint8 或 ﬂoat32。当传入函数时应 用中括号 [] 括来例如[img]

- channels: 同样用中括号括来它会告函数我们统幅图 像的直方图。如果入图像是灰度图它的值就是 [0]如果是彩色图像 的传入的参数可以是 [0][1][2] 它们分别对应着 BGR。

- mask: 掩模图像。统整幅图像的直方图就把它为 None。但是如 果你想统图像某一分的直方图的你就制作一个掩模图像并 使用它。

- histSize:BIN 的数目。也应用中括号括来

- ranges: 像素值范围常为 [0256] 


```python
img = cv2.imread('31.jpg',0) # 0 表示灰度图
hist = cv2.calcHist([img],[0],None,[256],[0,256])
hist.shape
```

```python
plt.hist(img.ravel(),256);
plt.show()
```

```python
img = cv2.imread('31.jpg')
color = ('b','g','r')
for i,col in enumerate(color): # 枚举
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    
plt.show()
```

# Mask掩码操作

```python
# 创建mask
mask = np.zeros(img.shape[:2],np.uint8)
mask[800:1000,200:8000] = 255
cv_show(mask,'mask')
```

```python
img = cv2.imread('31.jpg',0)
cv_show(img,'img')
```

```python
masked_img = cv2.bitwise_and(img,img, mask=mask) # 与 操作
cv_show(masked_img,'masked_img')
```

```python
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.subplot(221),plt.imshow(img,'gray')
plt.subplot(222),plt.imshow(mask,'gray')
plt.subplot(223),plt.imshow(masked_img,'gray')
plt.subplot(224),plt.plot(hist_full),plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```

```python
img = cv2.imread('31.jpg',0)
plt.hist(img.ravel(),256)
plt.show()
```

```python
equ = cv2.equalizeHist(img) # 均衡化函数
plt.hist(equ.ravel(),256)
plt.show()
```

```python
res = np.hstack((img,equ))
cv_show(res,'res')
```

# 自适应直方图均衡化

   
   --分小块做均衡化(小心噪音点)

```python
# 方法： 
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8,8))
```

```python
res_clahe = clahe.apply(img)
res = np.hstack((img,equ,res_clahe))

cv_show(res,'res') 
```
